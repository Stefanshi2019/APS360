### Part (a) Convolutional Network - 5 pt

Build a convolutional neural network model that takes the (224x224 RGB) image as input, and predicts the gesture
letter. Your model should be a subclass of nn.Module. Explain your choice of neural network architecture: how
many layers did you choose? What types of layers did you use? Were they fully-connected or convolutional?
What about other decisions like pooling layers, activation functions, number of channels / hidden units?

```python
kernel_size = [10, 5, 3]
class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        self.name = "CNN"
        self.conv1 = nn.Conv2d(3, 5, kernel_size[0])
        self.conv2 = nn.Conv2d(5, 10, kernel_size[1])
        self.conv3 = nn.Conv2d(10, 20,  kernel_size[2])
        self.pool = nn.MaxPool2d(2, 2)
        self.fc1_len = int((int((int((224 - kernel_size[0] + 1)/2) - kernel_size[1] + 1)/2) - kernel_size[2] + 1)/2)
        self.fc1 = nn.Linear(20 * self.fc1_len **2, 32)
        self.fc2 = nn.Linear(32, len(classes))

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = self.pool(F.relu(self.conv3(x)))
        x = x.view(-1, 20 * self.fc1_len**2)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        #x = x.squeeze(1) # Flatten to [batch_size]
        return x
```

I used 3 conv layers, with (3, 5, 10), (5, 10, 5), (10, 20, 3) input channels, output channels, and kernel sizes respectively, 1 maxpooling layer, and 2 fully connected layers. The total # neurons of the first fc layer is the # neurons of current layer, which is 20, times square of the size of previous layer, where in this case is calculated recursively using this formula W' = (W-F+2P) / S, where P=0, S=1, W = width of previous layer, F = kernel size. There are 32 input units to fc2, and length of classes, which is 9, as output. I used relu as Maxpool2d as pooling layers, relu as activation functions.

### Part (b) Training Code - 5 pt

Write code that trains your neural network given some training data. Your training code should make it easy
to tweak the usual hyperparameters, like batch size, learning rate, and the model object itself. Make sure
that you are checkpointing your models from time to time (the frequency is up to you). Explain your choice
of loss function and optimizer.

```python
import matplotlib.pyplot as plt
```

```python
def get_model_name(name, batch_size, learning_rate, epoch):
    path = "model_{0}_bs{1}_lr{2}_epoch{3}".format(name, batch_size, learning_rate, epoch)
    return path

def evaluate(net, loader):
  correct = 0
  total = 0
  for inputs, labels in loader:
    inputs = inputs.cuda()
    labels = labels.cuda()
    output = net(inputs)

    predictions = output.max(1, keepdim = True)[1]
    correct += predictions.eq(labels.view_as(predictions)).sum().item()
    total += inputs.shape[0]
  return correct/total

def train(net, batch_size = 128, learning_rate = 0.001, num_epochs = 20):
    # BCE doesnt seem to work
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.SGD(net.parameters(), lr=learning_rate, momentum=0.9)
    train_loader = torch.utils.data.DataLoader(train_set, batch_size=batch_size, num_workers=1, shuffle = True)
    val_loader = torch.utils.data.DataLoader(val_set, batch_size=batch_size, num_workers=1, shuffle = True)
    iters, losses, train_acc, val_acc = [], [], [], []
    print("training starts")
    for epoch in range(num_epochs):
        for inputs, labels in iter(train_loader):
            inputs = inputs.cuda()
            labels = labels.cuda()
            output = net(inputs)
            loss = criterion(output, labels)
            loss.backward()
            optimizer.step()
            optimizer.zero_grad()

        iters.append(epoch)
        losses.append(float(loss)/ batch_size)
        train_acc.append(evaluate(net, train_loader))
        val_acc.append(evaluate(net, val_loader))
      
        default_path = "/content/drive/My Drive/APS360/Lab3_results/"
        model_path = default_path + get_model_name(net.name, batch_size, learning_rate, epoch)
        torch.save(net.state_dict(), model_path)
        print(("Epoch {}: Train acc: {} |"+
                  "Validation acc: {}").format(
                      epoch,
                      train_acc[epoch],
                      val_acc[epoch]))
   
    plt.title("Training Curve")
    plt.plot(iters, losses, label="Train")
    plt.xlabel("Iterations")
    plt.ylabel("Loss")
    plt.show()

    plt.title("Training Curve")
    plt.plot(iters, train_acc, label="Train")
    plt.plot(iters, val_acc, label="Validation")
    plt.xlabel("Iterations")
    plt.ylabel("Training Accuracy")
    plt.legend(loc='best')
    plt.show()
```

### Part (c) “Overfit” to a Small Dataset - 5 pt

One way to sanity check our neural network model and training code is to check whether the model is capable
of “overfitting” or “memorizing” a small dataset. A properly constructed CNN with correct training code
should be able to memorize the answers to a small number of images quickly.

Construct a small dataset (e.g. just the images that you have collected). Then show that your model and

With a large batch size (e.g. the entire small dataset) and learning rate that is not too high, You should be

```python
def overfit(net, train_loader, batch_size = 64, learning_rate = 0.001, num_epochs = 30):
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.SGD(net.parameters(), lr=learning_rate, momentum=0.9)
    train_acc = []
    print("training starts")

    for epoch in range(num_epochs):
        for inputs, labels in iter(train_loader):
            inputs = inputs.cuda()
            labels = labels.cuda()
            output = net(inputs)
            loss = criterion(output, labels)
            loss.backward()
            optimizer.step()
            optimizer.zero_grad()
        acc = evaluate(net, train_loader)
        train_acc.append(acc)
        print("Epoch: {0}, Accuracy: {1}".format(epoch, train_acc[-1]))
        if acc==1.0:
          break
```

### 4. Transfer Learning [15 pt]
For many image classification tasks, it is generally not a good idea to train a very large deep neural network
model from scratch due to the enormous compute requirements and lack of sufficient amounts of training
data.

One of the better options is to try using an existing model that performs a similar task to the one you need
to solve. This method of utilizing a pre-trained network for other similar tasks is broadly termed **Transfer
Learning**. In this assignment, we will use Transfer Learning to extract features from the hand gesture
images. Then, train a smaller network to use these features as input and classify the hand gestures.

As you have learned from the CNN lecture, convolution layers extract various features from the images which
get utilized by the fully connected layers for correct classification. AlexNet architecture played a pivotal
role in establishing Deep Neural Nets as a go-to tool for image classification problems and we will use an
ImageNet pre-trained AlexNet model to extract features in this assignment.

The alexnet model is split up into two components: *alexnet.features* and *alexnet.classifier*. The

The neural network alexnet.features expects an image tensor of shape Nx3x224x224 as input and it will

Compute the AlexNet features for each of your training, validation, and test data. Here is an example code

### Part (b) - 3 pt
Build a convolutional neural network model that takes as input these AlexNet features, and makes a
prediction. Your model should be a subclass of nn.Module.

Explain your choice of neural network architecture: how many layers did you choose? What types of layers
did you use: fully-connected or convolutional? What about other decisions like pooling layers, activation

Here is an example of how your model may be called:

```python
# features = ... load precomputed alexnet.features(img) ...
# output = model(features)
# prob = F.softmax(output)

class AlexNet(nn.Module):
    def __init__(self):
        self.name = "alexnet"
        super(AlexNet, self).__init__()
        self.conv1 = nn.Conv2d(256, 512, 3)
        self.fc1 = nn.Linear(512*4*4, 32)
        self.fc2 = nn.Linear(32, 9)     

    def forward(self, x):
        x = F.relu(self.conv1(x))
        x = x.view(-1, 512*4*4)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return F.softmax(x, dim=1)
        
```

Answer: I used 1 convolutional layer, and 2 fully connected layers. The activation function is relu. There are 256 input channel, 512 output channels to the convolutional layer, and 512*4*4 hidden units in the linear layer, which is # neurons of current layer * # neurons of the previous conv layer. Since Alex Net already extracts features, I do not think there's a need for additional conv layers or maxpooling layers.

```python
# evaluate
def evaluate_own(net, loader):
  correct = np.zeros(9)
  total = np.zeros(9)
  for inputs, labels in loader:
    inputs = inputs.cuda()
    labels = labels.cuda()
    output = net(inputs)
    output = output.cpu()
    labels = labels.cpu()
    predictions = output.max(1, keepdim = True)[1].squeeze(1)
    labels = labels.numpy()
    predictions = predictions.numpy()
    #print("labels are ", labels)
    #print("predictions are\n", predictions)

    for i in range(len(labels)):
      if labels[i] == predictions[i]:
          correct[labels[i]] += 1
      # else:
      #     print("predicted ", classes[predictions[i]], ", should have been ", classes[labels[i]])
      total[labels[i]] += 1

  #print(correct)
  #print(len(loader))
  return correct/total
```
